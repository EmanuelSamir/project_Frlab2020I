```python
df_a = pd.read_csv("/tmp/xactual.dat", header = None, names=['t','x','y','z', ], sep = ',')#, usecols=columns_to_keep)

df_d = pd.read_csv("/tmp/xdeseado.dat", header = None, names=['t','x','y','z'], sep = ',')#, usecols=columns_to_keep)
i = 0
f, axs = plt.subplots(3, 1, figsize=(15,6))
coord = ['x','y','z']
low = [-0.8,0.7,1.3]
high = [-0.3, 1.6, 1.7]

for j in range(3):
    i = i + 1
    joint = coord[j]
    axs[j].plot(df_a['t'].values, df_a[joint].values, label = 'actual')
    axs[j].plot(df_d['t'].values, df_d[joint].values, label = 'deseado', color= 'r')
    axs[j].set(xlabel='tiempo (s)', ylabel='m', title=joint)
    axs[j].legend()
    axs[j].set_yticks(np.linspace(low[j], high[j], 5))
    axs[j].grid()
```
