```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_a = pd.read_csv("/tmp/qactual.dat", header = None, names=['t','q1','q2','q3','q4','q5','q6'], sep = ',')#, usecols=columns_to_keep)

df_d = pd.read_csv("/tmp/qdeseado.dat", header = None, names=['t','q1', 'q2','q3', 'q4', 'q5', 'q6'], sep = ',')#, usecols=columns_to_keep)
i = 0
f, axs = plt.subplots(3, 2, figsize=(12,16))
for j in range(3):
    for k in range(2):
        i = i + 1
        joint = 'q' + str(i)
        axs[j,k].plot(df_a['t'].values, df_a[joint].values, label = 'actual')
        axs[j,k].plot(df_d['t'].values, df_d[joint].values, label = 'deseado')
        axs[j,k].set(xlabel='t(seg)', ylabel='rad', title=joint)
        axs[j,k].legend()
```
